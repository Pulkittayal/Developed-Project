```python
x_data=df[["bedrooms","grade","sqft_above","sqft_living"]]
y_data=df["price"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.25,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
yhat=lm.predict(x_test)
```

```python
yhat[0:5]
```

```python
y_test[0:5]
```

```python
ax1=sns.distplot(y_test,label="Actual value",color="b",hist=False)
sns.distplot(yhat,label="Predicted value",color="r",hist=False,ax=ax1)
```

```python
lm.score(x_test,y_test)
```
