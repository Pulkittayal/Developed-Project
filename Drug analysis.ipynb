```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
```

```python
df=pd.read_csv(r"C:\Users\Pulkit\Desktop\Titanic kaggle\drug200.csv")
```

```python
df=pd.get_dummies(data=df,columns=["Sex","BP","Cholesterol"])
```

```python
df["Age"]=df["Age"].fillna(df["Age"].mean())
```

```python
df.head()
```

```python
x_data=df.drop("Drug",axis=1)
y_data=df["Drug"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.2,random_state=0)
z=[]
```

```python
def data_model(ml_model,x_train,x_test,y_train,y_test):
    model=ml_model
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print(confusion_matrix(y_test,y_pred))
    print(classification_report(y_test,y_pred))
    train_model= model.score(x_train,y_train)
                      
    test_model=accuracy_score(y_pred,y_test)
    
    print("The training accuracy is:",train_model)
    print("The testing accuracy is:",test_model)
    
    return z.append([ml_model,x_train,x_test,y_train,y_test])
```

```python
data_model(KNeighborsClassifier(),x_train,x_test,y_train,y_test)
```

```python
data_model(DecisionTreeClassifier(criterion="entropy"),x_train,x_test,y_train,y_test)
```

```python
data_model(LogisticRegression(),x_train,x_test,y_train,y_test)
```

```python
data_model(svm.SVC(kernel="rbf"),x_train,x_test,y_train,y_test)
```

```python
# so we see that DecisionTree is best alogrithm
```
